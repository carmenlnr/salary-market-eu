```python
# 1. Setup
import pandas as pd
```

```python
# 2. Conexion a MySQL 
import os
from dotenv import load_dotenv
from sqlalchemy import create_engine

# Cargamos las variables del .env
load_dotenv("../.env")

usuario = os.getenv("MYSQL_USER")
password = os.getenv("MYSQL_PASSWORD")
host = os.getenv("MYSQL_HOST")
puerto = os.getenv("MYSQL_PORT")
base_datos = os.getenv("MYSQL_DB")

# Creamos la conexion (engine) con SQLAlchemy
engine = create_engine(f"mysql+pymysql://{usuario}:{password}@{host}:{puerto}/{base_datos}")

print("Conexion creada correctamente")
```

```python
# 4. Cargamos los 3 datasets limpios
df_ofertas = pd.read_csv("../Data/ofertas_clean.csv")
df_salarios = pd.read_csv("../Data/ofertas_salarios_clean.csv")
df_hist = pd.read_csv("../Data/historico_clean.csv")

print("ofertas:", df_ofertas.shape)
print("salarios:", df_salarios.shape)
print("historico:", df_hist.shape)
```

```python
# 5. Subimos cada dataset a MySQL como una tabla
df_ofertas.to_sql("ofertas", con=engine, if_exists="replace", index=False)
df_salarios.to_sql("salarios", con=engine, if_exists="replace", index=False)
df_hist.to_sql("historico", con=engine, if_exists="replace", index=False)

print("Tablas creadas en MySQL: ofertas, salarios, historico")
```

```python
# Comprobamos que los datos se cargaron bien contando las filas

print(pd.read_sql("SELECT COUNT(*) AS total FROM ofertas", con=engine))
print(pd.read_sql("SELECT COUNT(*) AS total FROM salarios", con=engine))
print(pd.read_sql("SELECT COUNT(*) AS total FROM historico", con=engine))
```
